# Give one major difference in syntax that Python has with other languages such as C.

One major difference between Python and C is the way they handle code blocks. Python uses indentation while C uses curly braces {} to define code blocks.

# How does variable assignment differ in Python compared with other languages such as C?

In Python, it's not required to declare the type of a variable. There is no declaration needed, and you can freely reassign the type of a variable with a new value of a different type. In C, you must declare the type of a variable before using it. Once a variable is declared with a type, it can only hold values of that type.

# Do the assignment operators (+, -, *, /, %, **) work for all data types? Why or Why not?

Assignment operators work wth data types where the operation makes logical sense. **They do not work for all data types.** A mismatch between an operator and a data type will prompt Python to raise an error.

# How does the * operator differ from the ** operator?

The * operator is used for multiplication. The ** operator is used for exponentiation, which means you raise a number to the power of another number.
